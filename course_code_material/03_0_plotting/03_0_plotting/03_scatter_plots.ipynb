```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Download the titanic dataset from kaggle (https://www.kaggle.com/c/titanic/data):

```python
df = pd.read_csv("/tmp/train.csv")
print(df.shape)
df.head()
```

```python
plt.plot(df.Age, df.Fare)
plt.title('Fare vs Age')
plt.show()
```

```python
plt.scatter(df.Age, df.Fare)
plt.title('Fare vs Age')
plt.show()
```

```python
df.dropna(inplace=True, subset=["Fare", "Age"])
df.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df.Age.values.reshape(-1, 1), df.Fare.values.reshape(-1, 1))
x = np.arange(df.Age.min(), df.Age.max(), 0.1).reshape(-1, 1)
y = model.predict(x)
```

```python
plt.scatter(df.Age, df.Fare, label="data")
plt.scatter(x, y, label="best fit")
plt.title('Fare vs Age')
plt.legend()
plt.show()
```

```python
plt.scatter(df.Age, df.Fare, label="data", c="g")
plt.plot(x, y, label="best fit", c="r")
plt.title('Fare vs Age')
plt.legend()
plt.show()
```
