```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.1)
y = np.square(x)
plt.plot(x, y)
plt.title('$y=x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y2 = x ** 3
plt.figure(figsize=(12, 5))
plt.plot(x, y, label="$y=x^2$")
plt.plot(x, y2, label="y=x^3")
plt.title('powers of x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
plt.savefig("figure1.png")
```
